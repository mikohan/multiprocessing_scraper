```python
import csv
import requests
import random
import time
from multiprocessing import Pool
```

```python
def proxy():
    with open('proxy1.txt', 'r') as f:
        reader = csv.reader(f)
        proxy_list = list(reader)
        random.shuffle(proxy_list)
    for proxy in proxy_list:
        yield proxy[1] + ':' + proxy[2]
        
# def proxies():
#     f = open('proxylist.txt', 'r')
#     a = f.read().split('\n')
#     random.shuffle(a)
#     for i in a:
#         yield i
        
def get_html(urls):
    for ur in urls:
        url = 'http://httpbin.org/ip'
        prox = next(proxy())
        print(prox)
        proxies = {'https' : prox, 'http': prox}
        try:
            r = requests.get(url, proxies=proxies, timeout=5)
            print(r)
            return r.json()['origin']
        except:
            time.sleep(2)
            prox = next(proxy())
            
        
def get_html_ebay(urls):
    for ur in urls:
        url = 'http://httpbin.org/ip'
        prox = next(proxy())
        print(prox)
        proxies = {'https' : prox, 'http': prox}
        try:
            r = requests.get(url, proxies=proxies, timeout=5)
            print(r)
            return r.json()['origin']
        except:
            time.sleep(2)
            prox = next(proxy())
        
```
